Libraries

```python
import os
```

Get Protein Directories (Assumed Running in Parent Directory)

```python
directories = next(os.walk('.'))[1]
directories.remove(".ipynb_checkpoints")
print(directories)
```

Create Projections

```python
# Number of random projections you wish to generate
num_proj = 100
```

```python
for d in directories:
    # Identify protein structure in directory
    files = set([os.path.splitext(x)[0] for x in os.listdir(d)])
    protein = [x for x in files if ("EMD" in x)][0]
    p_map = protein + ".mrc"
    
    # Create projections folder if it does not already exist
    if not os.path.exists(os.path.join(d,"projections")):
        os.makedirs(os.path.join(d,"projections"))
        
    # Create (num_proj) projections from random orientations
    for i in range(num_proj):
        # Map Location 
        map_loc = "./" + os.path.join(d,p_map)
        #print(map_loc)
        # Output .hdf file
        out_loc = "./" + os.path.join(d,"projections","rand_" + str(i) + "_" + protein + ".hdf") #png
        #print(out_loc)
        # Create Projection Command String
        mk_proj = "~/Desktop/eman2/programs/e2project3d.py " + map_loc + " --outfile=" + out_loc + " --orientgen=rand:n=1 --sym=c3 --projector=standard --verbose=2"
        #print(mk_proj)
        
        #os.system(mk_proj) # Create Projection
        %run $mk_proj
        
        
        # Output .png file
        out_loc_png = "./" + os.path.join(d,"projections","rand_" + str(i) + "_" + protein + ".png") #jpg
        # Convert Projection from .hdf to .png Command String
        conv_proj = "~/Desktop/eman2/programs/e2proc2d.py " + out_loc + " " + out_loc_png
        #print(conv_proj)
        
        #os.system(conv_proj) # Convert Projection Format
        %run $conv_proj
```
