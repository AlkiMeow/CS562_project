Libraries

```python
import os
import shutil
```

Subsample from Main Dataset

```python
# Create Dataset Folder Structure if New
if not os.path.exists("Mini_Dataset"):
    os.makedirs("Mini_Dataset")
    
if not os.path.exists(os.path.join("Mini_Dataset",'images')):
    os.makedirs(os.path.join("Mini_Dataset",'images'))
    
if not os.path.exists(os.path.join("Mini_Dataset",'masks')):
    os.makedirs(os.path.join("Mini_Dataset",'masks'))
```

```python
# Number of Desired Samples in Mini Dataset
num_samples = 500
```

```python
# Get naming convention file format of samples
samples = os.listdir(os.path.join("Dataset",'images'))
sub_samples = [s[5:] for s in samples][:num_samples]
```

```python
# Perform Copy Over
for i, samp in enumerate(sub_samples):
    print("Now copying sample number " + str(i+1) + " with name: " + samp)
    
    orig_img = os.path.join("Dataset",'images', 'image'+samp)
    targ_img = os.path.join("Mini_Dataset",'images', 'image'+samp)
    shutil.copyfile(orig_img, targ_img)
    
    orig_msk = os.path.join("Dataset",'masks', 'mask'+samp)
    targ_msk = os.path.join("Mini_Dataset",'masks', 'mask'+samp)
    shutil.copyfile(orig_msk, targ_msk)
```
